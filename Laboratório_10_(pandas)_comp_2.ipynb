Neste exercício vamos importar, manipular e visualizar dados usando o módulo pandas.
O objetivo é desenhar um gráfico de dispersão analisando as áreas, os valores de condomínio e os preços de apartamentos no Rio de Janeiro, como pode ser visto na Figura 1.

Primeiramente, importe os módulos pandas, matplotlib.pyplot como pd, plt respectivamente.
Crie uma função chamada grafico cujo parâmetro de entrada e um inteiro n, seguem mais detalhes:

• Importe os dados que estão no arquivo dados.csv como um pd.DataFrame, esses dados podem ser encontrados em:
https://raw.githubusercontent.com/mvinoba/notebooks-for-binder/master/dados.csv

• Ordene os dados por ’area’ e define então um outro pd.DataFrame chamado subconjunto que contém apenas os dados de n apartamentos com a maior área.

• Numa figura, num eixo plote o gráfico de dispersão "Área vs valor do condomínio", onde os pontos representando os apartamentos de um bairro têm a mesma cor.

• Facultativamente, o tamanho de cada ponto pode refletir o preço de cada apartamento.

• O valor de retorno da função são as linhas do subconjunto que contêm o maior ou o menor preço do apartamento.

```python
import pandas as pd
import matplotlib.pyplot as plt

def grafico(n):
  '''n (int) -> df[0:n]'''
  link = 'https://raw.githubusercontent.com/mvinoba/notebooks-for-binder/master/dados.csv'
  dados = pd.read_csv(link)

  #ordenando os dados por area (descendente):
  copia = dados.sort_values('area', ascending=False).copy()

  #colocando no intervalo:
  subconjunto = copia[0:n]

  #plotando o gráfico:
  grupos_bairro = subconjunto.groupby('bairro')

  fig, ax = plt.subplots()

  tamanho = subconjunto['pm2'] / subconjunto['pm2'].max()
  for bairro, grupo in grupos_bairro:
      ax.scatter(grupo['area'], grupo['pm2'], s=(grupo['pm2']/100), alpha=0.5, label=bairro)

  ax.set_xlabel('Área (m²)')
  ax.set_ylabel('Preços (m²)')
  ax.set_title('Gráfico de Dispersão de Preços vs Área por Bairro')

  ax.legend()
  plt.show()


  maximo = dados[dados['pm2'] == dados['pm2'].max()]
  maximo
  minimo = dados[dados['pm2'] == dados['pm2'].min()]
  minimo
  print(maximo)
  print(minimo)
```

```python
grafico(200)
```

```python
grafico(100)
```
