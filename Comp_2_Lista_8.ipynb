#Questão 1. <br>
A função limaçon pode ser parametrizada da seguinte forma: <br>
r = r0 + cos(θ) <br>
x = r cos(θ) <br>
y = r sin(θ) <br>

Use esta definição para plotar a curva para r0 = 0.5, r0 = 1.0, r0 = 1.5 no intervalo [−4π, 4π].
Use pontos suficientes para que a curva esteja suave. Use cores diferentes, formas diferentes para cada linha e legendas para identificar cada curva. Adicione um título à figura e etiquetas para os eixos x e y.
Salve a figura em formato png.
Obs: a figura deve ter tamanho 1024x768.

```python
def limacon():
  r0 = [0.5, 1.0, 1.5]
  intervalo = np.linspace(-4*np.pi, 4*np.pi, 100) #50 pontos no intervalo de -4pi a 4pi
  fig, ax = plt.subplots()
  for i in r0:
    r = i + np.cos(intervalo)
    x = r * np.cos(intervalo)
    y = r * np.sin(intervalo)
    ax.plot(x, r, '*',linewidth=2, label=f'r de {i}')
    ax.plot(x, x, 'v', linewidth=2, label=f'x de {i}')
    ax.plot(x, y, '--', linewidth=2, label=f'y de {i}')
    plt.xlabel('Intervalo de -4pi a 4pi')
    plt.ylabel('Valores da parametrização')
    ax.legend(loc='upper right')

  plt.title('Limaçon Parametrizada')
  plt.show()
  plt.savefig('Gráfico.png', format='png', dpi=100) #dpi=100 -> para salvar na resolução de 1024x768
  #ideal colocar o endereço de destino a ser salvo
```

```python
limacon()
```

#Questão 2. <br>
Use a matplotlib e numpy para plotar as funções
f(x) = e^(−x/10).sin(πx) e g(x) = xe^(x/3) no intervalo [0, 10]. <br>
Inclua etiquetas para os eixos x e y, e uma legenda identificando cada linha. <br>
Cada linha deve ter um formato e uma cor. Salve a figura em formato jpg.

```python
x = np.linspace(0, 10, 100) #intervalo de 0 a 10 com 100 pontos
fx = np.exp(-x / 10) * np.sin(np.pi * x) #função f(x)
gx = x * np.exp(x / 3) #função g(x)

plt.figure(figsize=(10, 7)) #criando a imagem

plt.plot(x, fx, 'b-', label='f(x) = e^(-x/10) * sin(πx)')
plt.plot(x, gx, 'r--', label='g(x) = xe^(x/3)')

plt.title('Funções f(x) e g(x)') #titulo
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.savefig('funcoes_plot.jpg', format='jpg') #salbando a figura no formato jpg
plt.show() #exibindo o gráfico
```

Carregue o arquivo nasa-giss.txt utilizando a função numpy.loadtxt passando como parâmetro o caminho do arquivo. <br>
Esta função irá ler o arquivo e criar uma matriz com 3 colunas. A primeira coluna são os anos, a segunda são os valores brutos de variação da temperatura e a terceira coluna são os valores de variação suavizados. <br>
Você deve plotar a variação bruta e a variação suavizada por ano. <br>
A variação bruta deve ter a cor azul, ser tracejada e ter a largura da linha igual a 1 pixel.  <br>
A variação suavizada deve ter a cor verde se for maior ou igual a 0 e vermelha caso contrário. <br>
Além disso a linha deve ter largura igual a 3 pixels e ser contínua. <br>
Ambos eixos devem possuir etiquetas e as curvas precisam ter legendas. Ao fim dê um título à imagem e salve em formato jpg.

```python
local = '/content/drive/MyDrive/Comp 2/nasa-giss.txt'
dados = np.loadtxt(local, dtype=float, delimiter=None)
#1 - anos
anos = dados[:,0]
#2 - valores brutos de variação
brutos = dados[:,1]
#3 - valores de variação suavizados
suavizados = dados[:,2]

plt.figure(figsize=(10, 7)) #criando a imagem

plt.plot(anos, brutos, 'b--', linewidth=1, label='valores brutos')
plt.plot(anos, suavizados, ('g-' if (suavizados >= 0.0).all() else 'r-'), linewidth=3, label='valores suavizados')


plt.legend()
plt.title('Variação no aumento da temperatura global')
plt.savefig('variacao.jpg', format='jpg')
plt.xlabel('Anos')
plt.ylabel('Valores')
plt.show()
```
